### Exercise 5: Subplot Layout

```python
import matplotlib.pyplot as plt

# Data for all plots
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# First subplot: Line chart
axs[0, 0].plot(x, y, color='blue')
axs[0, 0].set_title('Line Chart')

# Second subplot: Bar chart
axs[0, 1].bar(categories, values, color=['red', 'green', 'blue', 'orange'])
axs[0, 1].set_title('Bar Chart')

# Third subplot: Scatter plot
axs[1, 0].scatter(x, y, color='purple')
axs[1, 0].set_title('Scatter Plot')

# Fourth subplot: Pie chart
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
axs[1, 1].set_title('Pie Chart')
axs[1, 1].axis('equal')  # Ensure pie is circular

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plots
plt.show()
```
